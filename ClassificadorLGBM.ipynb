```python
preditoras = train.iloc[:, 0:524]
target = train.iloc[:, 524]
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(preditoras, target, test_size=0.3, random_state=42)
```

```python

modelo = LGBMClassifier(random_state=42, n_jobs=-1)
modelo.fit(preditoras, target)


seletor = SelectFromModel(estimator=modelo, prefit=True,threshold= 0.01)
cols_selecionadas = preditoras.columns[seletor.get_support()]

x_train_selecionado = preditoras[cols_selecionadas]
x_test_selecionado  = test[cols_selecionadas]

print(f"Shape treino original: {preditoras.shape}")
print(f"Shape treino após seleção: {x_train_selecionado.shape}")
print(f"Shape teste após seleção: {x_test_selecionado.shape}")
print(f"Número de features selecionadas: {len(cols_selecionadas)}")
```

```python
cols = seletor.get_support(indices=True)
x_treino_selecionado = x_treino.iloc[:, cols]
x_teste_selecionado = x_teste.iloc[:, cols]
```

```python
classificador_lgbm = lgb.LGBMClassifier()
```

```python
print(classificador_lgbm.get_params())
```

```python
grid_parametros = {
    'n_estimators': [200, 300, 500, 800],
    'max_depth': [5, 8, 12, 15],
    'num_leaves': [10, 20, 30, 50, 70],
    'learning_rate': [0.005, 0.01, 0.02, 0.05, 0.1],
    'subsample': [0.6, 0.7, 0.8, 0.9, 1.0],
    'colsample_bytree': [0.6, 0.7, 0.8, 0.9, 1.0],
    'reg_alpha': [0.0, 0.1, 0.5, 1.0, 2.0],
    'reg_lambda': [0.0, 0.1, 0.5, 1.0, 2.0],
    'min_child_samples': [5, 10, 20, 30],
    'min_child_weight': [0.001, 0.01, 0.1]
}

#RandomizedSearchCV
randomLGBM = RandomizedSearchCV(
    estimator=LGBMClassifier(),
    param_distributions=grid_parametros,
    n_iter=80,
    cv=3,
    scoring='accuracy',
    n_jobs=-1,
    random_state=120,
    verbose=2
)


randomLGBM.fit(x_train_selecionado, target)
treinos_lgbm = pd.DataFrame(randomLGBM.cv_results_)

print(f"Acurácia em Treinamento: {randomLGBM.best_score_ :.2%}")
print("")
print(f"Hiperparâmetros Ideais: {randomLGBM.best_params_}")
print("")
```

```python
#Novo classificador
from lightgbm import LGBMClassifier

clf = LGBMClassifier(
    subsample=1.0,
    reg_lambda=0.1,
    reg_alpha=0.0,
    num_leaves=70,
    n_estimators=500,
    min_child_weight= 0.001,
    min_child_samples= 10,
    max_depth=5,
    learning_rate=0.1,
    colsample_bytree=0.9,
    objective='binary',
    random_state=42,
    n_jobs=-1
)

clf.fit(x_train_selecionado, target)
```

```python

num_folds = 5
seed = 3
kfold = KFold(n_splits=num_folds, shuffle=True, random_state=seed)

modelo_final = LGBMClassifier(**randomLGBM.best_params_)
resultado = cross_val_score(modelo_final, x_train_selecionado, target, cv=kfold)

print("Acurácia CV: %.2f%% (%.2f%%)" % (resultado.mean()*100, resultado.std()*100))
```

```python
y_pred = clf.predict(x_test_selecionado)
y_pred_labels = lb.inverse_transform(y_pred)

submissao = test[['id']].copy()
submissao['target'] = y_pred_labels
submissao.to_csv('submission.csv', index=False)


```
